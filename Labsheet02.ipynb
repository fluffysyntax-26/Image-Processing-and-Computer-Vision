```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Labsheet-2 Image Enhancement Techniques

Objective: Apply Contrast and Brightness enhancement<br>
Task: Use Histogram equalisation and contrast stretching<br>
Activity: Compares results on different image datasets

```python
def contrast_stretching(img): 
    #min-max normalisation
    min_val = np.min(img)
    max_val = np.max(img)

    stretched = (img-min_val) * (255/(max_val - min_val))
    return stretched.astype(np.uint8)
```

```python
# Load Image (change file name for other datasets)
image_path = "lena.jpg" 
img = cv2.imread(image_path, 0) # 0 to read as grayscale
```

```python
# Histogram equalitsation
hist_eq = cv2.equalizeHist(img)

# Contrast Stretching
contrast_stretched = contrast_stretching(img)
```

```python
# Plotting Results
titles = ['Original Image', "Histogram Equalisation", "Constrast Stretching"]
images = [img, hist_eq, contrast_stretched]

plt.figure(figsize=(12, 6))
for i in range(3): 
    plt.subplot(1, 3, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i] + " Histogram")

plt.show()
    
```

```python
plt.figure(figsize=(12, 6))
for i in range(3): 
    plt.subplot(1, 3, i+1)
    plt.hist(images[i].ravel(), 256, [0, 256])
    plt.title(titles[i] +" Histogram")

plt.show()
```
